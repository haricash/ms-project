```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def salpeter(m):
    """
        Calculates the Salpeter IMF
    """
    if m >= 0.1 and m <= 125 :
        return 1.35

def miller_scalo(m):
    """
        Calculates the Miller & Scalo IMF
    """
    if m >= 0.1 and m < 1 :
        return 0.25
    elif m >= 1 and m < 2 :
        return 1
    elif m >= 2 and m < 10 :
        return 1.3
    elif m >= 10 and m <= 125 :
        return 2.3
```

```python
salpeter = np.vectorize(salpeter)
miller_scalo= np.vectorize(miller_scalo)
```

```python
constant = 1
```

```python
mass = np.linspace(start = 0.1,
                   stop = 125,
                   num = 1000)

salpeter_nos = constant*np.power(mass, -(1+salpeter(mass)))
miller_scalo_nos = constant*np.power(mass, -(1+miller_scalo(mass)))
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,6))


ax[0].plot(mass, salpeter_nos)
ax[1].plot(mass, miller_scalo_nos)

ax[0].set_ylabel('Relative numbers')
ax[0].set_xlabel("Mass in $M_{\odot}$")
ax[0].set_title("Salpeter IMF")

ax[1].set_ylabel('Relative numbers')
ax[1].set_xlabel("Mass in $M_{\odot}$")
ax[1].set_title("Miller and Scalo IMF")

ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[0].grid()

ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[1].grid()
```
